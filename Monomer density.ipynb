```python
#plot monomer density as function of height
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
import bisect

z_size = 100
n_polymers = 100

#initialize empty list to bin height values:
height = []
height.append(0)
    
#create bins:
h=0
while h < (z_size-1):
    h = h + 1
    height.append(h)
       
Z = []
        
#Get z coordinates for each monomer
for i in np.arange(0, n_polymers):
    #read in each csv file and add the z coordinates to list Z
    df = pd.read_csv(f'polymer1_{i}.csv')
    for index, row in df.iterrows():
        Z.append(row['z'])

#sort the list 
#Z.sort()

def count(Z, a, b):
     
    # x for x in list1 is same as traversal in the list
    # the if condition checks for the number of numbers in the range 
    # l to r 
    # the return is stored in a list
    # whose length is the answer
    return len(list(x for x in Z if a <= x < b))

density = []
zs = z_size

for k in range (0, zs):
    e = k
    f = k+1
    c = count(Z, e, f)
    density.append(c)

plt.plot(height,density)
plt.xlabel('Height (z)')
plt.ylabel('Monomer density')
plt.show()    

da = {'z' : height, 'density': density}
df = pd.DataFrame(data=da)
df.to_csv("Density_trial_1.csv")
    

```
